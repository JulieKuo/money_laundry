```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
info = pd.read_csv("data/feat_gen/info.csv")
info["date"] = pd.to_datetime(info["date"])
print(info.shape)
info.head(3)
```

```python
cdtx = pd.read_csv("data/feat_gen/cdtx.csv")
cdtx["date"] = pd.to_datetime(cdtx["date"])
print(cdtx.shape)
cdtx.head(3)
```

```python
ccba = pd.read_csv("data/feat_gen/ccba.csv")
ccba["byymm"] = pd.to_datetime(ccba["byymm"])
print(ccba.shape)
ccba.head(3)
```

```python
info = info.drop("total_asset", axis = 1)
```

```python
cdtx = cdtx.drop("amt", axis = 1)
df = pd.merge(info, cdtx, on = ["cust_id", "date"], how = "left")
df.head()
```

```python
df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month
df["day"] = df["date"].dt.day
df = df.drop("date", axis = 1)
df.head()
```

```python
ccba = ccba[['cust_id', 'byymm', 'amt_over']].copy()
ccba["year"] = ccba["byymm"].dt.year
ccba["month"] = ccba["byymm"].dt.month
ccba = ccba.drop("byymm", axis = 1)
ccba.head()
```

```python
df = pd.merge(df, ccba, on = ["cust_id", "year", "month"], how = "left")
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[:, 7:] = df.iloc[:, 7:].fillna(0)
```

```python
df.isnull().sum()
```
