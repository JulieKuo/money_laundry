```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cdtx = pd.read_csv("data/inverse/cdtx.csv")
cdtx["date"] = pd.to_datetime(cdtx["date"])
print(cdtx.shape)
print("cust_id:", cdtx["cust_id"].nunique())
cdtx.head()
```

```python
df = cdtx.groupby(["cust_id", "date"]).agg({"amt": ["sum", "count"], "country": "count", "cur_type": "count"})
df.columns = ["amt", "amt_ct", "country_ct", "cur_type_ct"]
df = df.reset_index()
df["country"] = df.eval("country_ct != 1").astype(int)
df["cur_type"] = df.eval("cur_type_ct != 1").astype(int)
df
```

```python
df.to_csv("data/feat_gen/cdtx.csv", index = False)
```
